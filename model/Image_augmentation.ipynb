```python
import matplotlib.pyplot as plt
import numpy as np
import os
import random
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
%matplotlib inline
```

```python
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 10, figsize=(20,20))
    axes = axes.flatten()
    for img, ax in zip( images_arr, axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
gen = ImageDataGenerator( rotation_range=0.1, brightness_range=(0.9,1.1), zoom_range=0.1,
    horizontal_flip=True)
```

```python
ripe_img_paths= os.listdir('Training/Ripe/')
medium_img_paths= os.listdir('Training/Medium/')
not_ripe_img_paths= os.listdir('Training/Not_Ripe/')
```

```python
for path in ripe_img_paths:
    image_path = 'Training/Ripe/' + path
    image = np.expand_dims(plt.imread(image_path),0)
    aug_iter = gen.flow(image,save_to_dir='augmented/Ripe',save_format='png')
    aug_images = [next(aug_iter)[0].astype(np.uint8) for i in range(8)]
```

```python
for path in medium_img_paths:
    image_path = 'Training/Medium/' + path
    image = np.expand_dims(plt.imread(image_path),0)
    aug_iter = gen.flow(image,save_to_dir='augmented/Medium',save_format='png')
    aug_images = [next(aug_iter)[0].astype(np.uint8) for i in range(10)]
```

```python
for path in not_ripe_img_paths:
    image_path = 'Training/Not_Ripe/' + path
    image = np.expand_dims(plt.imread(image_path),0)
    aug_iter = gen.flow(image,save_to_dir='augmented/Not_Ripe',save_format='png')
    aug_images = [next(aug_iter)[0].astype(np.uint8) for i in range(1)]
    
```
